```python
from google_play_scraper import search
from pprint import pprint

def get_app_data(query, num_results=10000):
    """
    Retrieves app data from Google Play and returns a pandas DataFrame.
    Args:
        query: The search query to use (e.g., empty string for all apps, or category name).
        num_results: The number of apps to retrieve.
    Returns:
        A pandas DataFrame containing app information, or None if an error occurs.
    """
    try:
        search_results = search(
            query,
            lang="en",  # Language for the search
            country="ph",  # Country for the search
        )
        
        # Check if search_results is a tuple (results, continuation_token) or just results
        if isinstance(search_results, tuple):
            results = search_results[0]
        else:
            results = search_results
        
        # Extract specific information and limit results to num_results
        extracted_data = []
        for app in results[:num_results]:
            extracted_data.append({
                "package_name": app.get("appId", "N/A"),
                "app_name": app.get("title", "N/A"),
                "category": query  # Since we're searching for "games"
            })
        
        # Try to convert to DataFrame
        try:
            import pandas as pd
            df = pd.DataFrame(extracted_data)
            return df
        except ImportError:
            print("Pandas is not installed or not working. Returning list of dictionaries instead.")
            return extracted_data
        
    except Exception as e:
        print(f"Error occurred during scraping: {e}")
        return None

# Retrieve app data using the get_app_data function
query = "audio"
app_data = get_app_data(query)

if app_data is not None:
    if isinstance(app_data, pd.DataFrame):
        print(f"Retrieved {len(app_data)} apps")
        print("First 5 apps:")
        print(app_data.head())
    else:
        print(f"Retrieved {len(app_data)} apps")
        print("First 5 apps:")
        for app in app_data[:5]:
            pprint(app)
else:
    print("An error occurred while retrieving app data.")
```
